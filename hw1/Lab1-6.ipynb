6. Реализуем функцию-генератор, которая генерирует 10 целых чисел и используем ее для построения массива

````python
def generator(start, step=1):
    for v in np.arange(start, (start) + 10 * step, step, dtype=np.int64):
        yield v
m = np.fromiter(generator(2, 2), dtype=np.int64)
m
````

7. Создайте массив случайных чисел размером 10 и отсортируйте его

````python
def random_generator(low=0, up=100):
    for _ in range(10):
        yield np.random.randint(low, up)
m = np.array([*random_generator()])
m.sort()
m
    
````

8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
m = np.array([[*random_generator()], [*random_generator()]])
r = np.sqrt(m[0]**2 + m[1]**2)
t =np.arctan2(m[1], m[0])
t = np.where(t < 0, t + 2 * np.pi, t)
mm = np.array([r, t])
mm
````

````python
import pandas as pd
````

````python
df = pd.read_csv("student_perfomance.csv")
df.head(5)
````

1. Посчитайте среднее кол-во часов уделяемое на сон студентами

````python
print(df["Sleep_Hours"].mean())
````

2. Сколько процентов студентов посещают внеклассовые активности?

````python
print(df['Extracurricular_Activities'].value_counts().iloc[0] / np.float128(df.shape[0]) * 100)
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
print("Правда" if 
      df[(df["Gender"] == "Female")]["Exam_Score"].mean() > 
      df[(df["Gender"] == "Male")]["Exam_Score"].mean()
      else "Ложь")
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
x = df["Sleep_Hours"].quantile(0.25)
y = df["Sleep_Hours"].quantile(0.75)
print(x, y)
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python

print("Правда" if 
      df[df["Sleep_Hours"] >= y]["Exam_Score"].min() > df[df["Sleep_Hours"] <= x]["Exam_Score"].max()
      else "Ложь")
````

6. Влияет ли физическая активность на результаты экзамена?

````python
print("Правда" if 
      df["Exam_Score"].corr(df["Physical_Activity"], "pearson") > 0
      else "Ложь")
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
a = df[(df["Parental_Education_Level"] == "College") | (df["Parental_Education_Level"] == "Postgraduate")]
b = df.drop(a.index)
a["Exam_Score"].min() > b["Exam_Score"].max()
print("Правда" if 
      a["Exam_Score"].min() > b["Exam_Score"].max()
      else "Ложь")
````

8. Как влияет участие родителей на результаты экзаменов

````python
df.groupby("Parental_Involvement")["Exam_Score"].describe()
cat_type = pd.CategoricalDtype(['Low', 'Medium', 'High'], ordered=True)
df["Parental_Involvement_code"] = pd.Categorical(df["Parental_Involvement"], dtype=cat_type).codes
df["Exam_Score"].corr(df["Parental_Involvement_code"], "spearman") > 0
print("Правда" if 
      df["Exam_Score"].corr(df["Parental_Involvement_code"], "spearman") > 0
      else "Ложь")
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python

print("Правда" if 
      df[df["Family_Income"] == "High"]["Exam_Score"].min() > 
      df[df["Family_Income"].isin(["Medium", "Low"])]["Exam_Score"].max()
      else "Ложь")
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с низким доходом и отсутствием доступа к интернету

````python
cat = df[(df["Internet_Access"] == "No") & (df["Family_Income"] == "Low")]
print((cat["Exam_Score"] > 70).value_counts().iloc[0] / cat.shape[0])
````
